```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('COVID19PHforMappingModified_31-03-2020_20h23m51s.csv')
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
Map_cluster = folium.Map(location=[0,0], height=600, width=950, tiles="OpenStreetMap",
                   zoom_start=1.5)

marker_cluster = MarkerCluster().add_to(Map_cluster)

for i in np.arange(len(df)):
    
    folium.CircleMarker([df['hosp_lat'][i], df['hosp_long'][i]], 
                        radius=2.5, color='black', clustered_marker = True, 
                        popup=folium.Popup('Admitted at: {0}'.format(df['hospital'].values[i]), 
                                           parse_html=True)).add_to(marker_cluster)
    
Map_cluster
```

```python
Map_cluster = folium.Map(location=[12.8797,121.7740], height=580, width=650, tiles="OpenStreetMap",
                   zoom_start=5)

marker_cluster = MarkerCluster().add_to(Map_cluster)

for i in np.arange(len(df)):
    lat = df['hosp_lat'][i]
    lon = df['hosp_long'][i]
    folium.CircleMarker([df['hosp_lat'][i], df['hosp_long'][i]], 
                        radius=2.5, color='black', clustered_marker = True, 
                        popup=folium.Popup('Admitted at: {0}'.format(df['hospital'].values[i]), 
                                           parse_html=True)).add_to(marker_cluster)
Map_cluster
```
